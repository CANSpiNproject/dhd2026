```python
# create /notebook_output folder

if not (os.path.isdir(os.path.join('notebook_output', 'merged_annotations'))):
    os.makedirs(os.path.join('notebook_output', 'merged_annotations'))
```

```python
# merge process

annotation_filenames: List[str] = [
    'DEU19_001_2-3-5.tsv',
    'DEU19_030_1-1-1.tsv',
    'DEU19_058_14.tsv'
]

for filename in annotation_filenames:
  
    # get filename, paths and load date into dataframes
    tsv_filename_llpro: str = filename
    tsv_filename_cs1: str = filename

    tsv_filepath_llpro: str = os.path.join('source_files', 'llpro-data', tsv_filename_llpro)
    tsv_filepath_cs1: str = os.path.join('source_files', 'cs1-data', tsv_filename_cs1)

    df_llpro: pd.DataFrame = pd.read_csv(filepath_or_buffer=tsv_filepath_llpro, delimiter='\t')
    df_cs1: pd.DataFrame = pd.read_csv(filepath_or_buffer=tsv_filepath_cs1, delimiter='\t')
  
    # unify df_cs1 and df_llpro step 1: delete linebreaks and change tokenization in df_cs1
    df_cs1_unified: pd.DataFrame = df_cs1[(~df_cs1['Token'].str.contains(r'\\n')) & (~df_cs1['Token'].str.contains(r' '))].reset_index(drop=True)
  
    # unify df_cs1 and df_llpro step 2: file specific unification measures;
    # this step is necessary because the tokenization in llpro and in gitma-canspin (which is used to create the cs1 tsv files) is done with different tokenizers
    df_cs1_unified = adjust_cs1_df(df=df_cs1_unified, text=filename)
    
    # check equality
    shortest_df_length: int = len(df_llpro.index) if len(df_llpro.index) < len(df_cs1_unified.index) else len(df_cs1_unified.index)
    for i in range(0, shortest_df_length):
        llpro_token: str = list(df_llpro.loc[[i]].text)[0]
        cs1_token: str = list(df_cs1_unified.loc[[i]].Token)[0]
        if llpro_token != cs1_token:
            print(f'{i} --- {llpro_token} --- {cs1_token}')
            raise ValueError('Both dataframes are not equal in terms of textual token and indices.')
        
    # create a joined dataframe;
    # this new dataframe gets the following columns:
    # i, text, lemma, event_id, event_label, correct_event_nonevent_prediction, Tag, Annotation_ID, Multi_Token_Annotation, main_tags_in_events, main_tags_in_non_events, reference_to_diegesis
    joined_df: pd.DataFrame = pd.concat(
        [
            df_llpro, 
            df_cs1_unified.Tag, 
            df_cs1_unified.Annotation_ID, 
            df_cs1_unified.Multi_Token_Annotation
        ],
        axis=1
    )
    joined_df = joined_df.drop(
        columns=[
            'orig', 
            'orig_offset', 
            'is_sent_start', 
            'is_para_start', 
            'is_section_start', 
            'pos', 
            'tag', 
            'morph', 
            'dep', 
            'head', 
            'speech', 
            'entity', 
            'character', 
            'coref_clusters', 
            'scene_id', 
            'scene_label'
        ]
    )
    joined_df = joined_df.reindex(
        joined_df.columns.tolist() + \
            [
                'main_tags_in_events',
                'main_tags_in_non_events', 
                'reference_to_diegesis'
            ],
        axis=1
    )
    joined_df.insert(
        5, 
        'correct_event_nonevent_prediction', 
        ' '
    )
    
    # export joined_df to csv
    joined_df.to_csv(os.path.join('notebook_output', 'merged_annotations', f'{filename[:-4]}.csv'))
```
